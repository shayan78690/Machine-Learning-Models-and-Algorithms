```python
# problem with the above plot interpretation
# you understand the relationship among features together
# but you don't understand how much >> lack  of quantitative measure
# so we study VIF
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# it measures the variance of one independent variable with respect to another independent variable
```

```python
vif = pd.DataFrame()
vif['Feature'] = df.columns
```

```python
vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(len(df.columns))]
```

```python
vif
```

```python
# if vif > 10 drop the features one by one
# highest is longitude we will drop this and thenm again calculate vif then again drop the highest one
```

```python
df1 = df.copy()
```

```python
df1
```

```python
df1.drop('Longitude', axis = 1, inplace = True)
```

```python
df1
```

```python
vif = pd.DataFrame()
vif['Feature'] = df1.columns
vif['VIF'] = [variance_inflation_factor(df1.values, i) for i in range(len(df1.columns))]
```

```python
vif
```

```python
df1.drop('AveRooms', axis = 1, inplace = True)
```

```python
vif = pd.DataFrame()
vif['Feature'] = df1.columns
vif['VIF'] = [variance_inflation_factor(df1.values, i) for i in range(len(df1.columns))]
```

```python
vif
```

```python
df1.drop('Latitude', axis = 1, inplace = True)
```

```python
vif = pd.DataFrame()
```

```python
vif['Features'] = df1.columns
vif['VIF'] = [variance_inflation_factor(df1.values, i) for i in range(len(df1.columns))]
```

```python
vif
```

```python
# now all the features we have vif below 10 so we can build the model
```

```python
df1
```

```python
X = df1.iloc[:, :-1] # all the rows except target variable
y = df1.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Scaling(Optional)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Performance
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score
```

```python
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```

```python
r2_score = r2_score(y_test, y_pred)
```

```python
ar2_score = 1 - (1 - r2_score) * (len(y_test) - 1) / ((len(y_test) - X_test.shape[1] - 1))
```

```python
ar2_score
```

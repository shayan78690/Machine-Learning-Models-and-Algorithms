```python
data = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
```

```python
data.head()
```

```python
data['target'] = diabetes.target
```

```python
data
```

```python
# EDA
# Data Cleaning
# Data Preparation
# Feature Engineering
# but in our case data is already cleaned
```

```python
# simple linear regression
# only one X and one y
```

```python
# but here we have multiple
# let's select one
X = data[['bmi']] # one independent variable
y = data['target']
```

```python
X
```

```python
y
```

```python
# train-test-split
# Split arrays or matrices into random train and test subsets.
# Quick utility that wraps input validation, next(ShuffleSplit().split(X, y)), and application to input data into a single call for splitting (and optionally subsampling) data into a one-liner.
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
# actually it is randomly splitting for not changing the value
```

```python
# random_state when fixed helps us to reprodice the same result again and again
```

```python
# scaling(optional)
```

```python
# model training
from sklearn.linear_model import LinearRegression
```

```python
# sklearn is the library
# linear_model is the module
# LinearRegression is the class
```

```python
model = LinearRegression()
```

```python
model
```

```python
X_train
```

```python
y_train
```

```python
# training of model
model.fit(X_train, y_train)
# at the backend it will use gradient descent to give optimal coefficient
```

```python
# this model traning will give you two thing intercept and coeffiecient
```

```python
model.coef_ # attribute of the model
```

```python
model.intercept_
```

```python
# finally you have built the model
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# visulaize the result
plt.scatter(X_test, y_test, color = 'black', label = 'Actual data')
plt.plot(X_test, y_pred, color = 'blue', linewidth = 3, label = 'Predicted Linear Regression Line')
plt.xlabel("BMI")
plt.ylabel("1 Year Disease Progression Target")
plt.title("Linear Regression on Linear Data")
plt.legend()
plt.show()
```

```python
# model evalutaion
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

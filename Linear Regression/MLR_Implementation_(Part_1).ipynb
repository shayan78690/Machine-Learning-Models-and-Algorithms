```python
# Problem Statement >> To predict the disease progression one year after the baseline
```

```python
diabetes.feature_names
```

```python
data = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
```

```python
data.head()
```

```python
data['target'] = diabetes.target
```

```python
data
```

```python
# EDA and Data Preparation
```

```python
# Division
X = data.drop('target', axis = 1)
y = data['target']
```

```python
X
```

```python
y
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
X_test
```

```python
# model training
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_train.shape
```

```python
# because there are 10 columns so there will be 10 coeffiecients
model.coef_
```

```python
model.intercept_
```

```python
# interpretation
X_train
# age has coeff -30
# one unit change in age the sugar level decreases by -30 units on an aveerage where all other features are kept constant
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("The mean squared erroe is ", mean_squared_error(y_test, y_pred))
```

```python
print("The r2 score is ", r2_score(y_test, y_pred))
```

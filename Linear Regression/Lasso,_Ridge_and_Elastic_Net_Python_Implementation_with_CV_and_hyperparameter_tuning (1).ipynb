```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('mpg')
```

```python
df
```

```python
df.drop("name", axis=1, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
#since we have not done outlier treatment, then better idea would be to replace the missing value with median

df['horsepower'].median()
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.dtypes
```

```python
df['origin'].value_counts()
```

```python
#data encoding
df["origin"] = df["origin"].map({"usa":1, "japan":2, "europe":3})
```

```python
# df["origin"] = df["origin"].astype(int)
```

```python
df.dtypes
```

```python
#separate into X and y
```

```python
X = df.drop('mpg', axis=1)
y = df['mpg']
```

```python
X
```

```python
y
```

```python
#train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#simple linear regression model
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
```

```python
regression_model
```

```python
regression_model.fit(X_train, y_train)
```

```python
regression_model.coef_
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {regression_model.coef_[i]}")
```

```python
#observations:
#coefficients are relatively smaller, if one independent variable changes
#there will be not much difference in prediction
#This is sometime called as smoother model

#these features might not be contributing in model training
```

```python
from sklearn.metrics import r2_score

y_pred_linear = regression_model.predict(X_test)
r2_linear = r2_score(y_test, y_pred_linear)
print(f"The R square of linear regression {r2_linear}")
```

```python
#regularised model
#ridge regression
```

```python
from sklearn.linear_model import Ridge
ridge_regression_model = Ridge(alpha=0.1)
ridge_regression_model
#in practical implementation lambda is alpha
```

```python
ridge_regression_model.fit(X_train, y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {ridge_regression_model.coef_[i]}")
```

```python
#for ridge regression evaluation
y_pred_ridge = ridge_regression_model.predict(X_test)
r2_ridge = r2_score(y_test, y_pred_ridge)
print(f"The R square of Ridge regression {r2_ridge}")
```

```python
#we dont see much variation in coeff of ridge regression as compared to linear regression
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_regression_model = Lasso(alpha = 0.5)
lasso_regression_model
```

```python
lasso_regression_model.fit(X_train, y_train)
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {lasso_regression_model.coef_[i]}")
```

```python
#three features coefficient is 0, lasso helps in feature selection
```

```python
y_pred_lasso = lasso_regression_model.predict(X_test)
r2_lasso = r2_score(y_test, y_pred_lasso)
print(f"The R square of Lasso regression {r2_lasso}")
```

```python
#Elastic net

from sklearn.linear_model import ElasticNet
elastic_net_model = ElasticNet(alpha = 1, l1_ratio=0.5)
elastic_net_model.fit(X_train, y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {elastic_net_model.coef_[i]}")
```

```python
y_pred_elastic = elastic_net_model.predict(X_test)
r2_elastic = r2_score(y_test, y_pred_elastic)
print(f"The R square of Elastic regression {r2_elastic}")
```

```python
#regularisation with crossvalidation
from sklearn.linear_model import LassoCV
lasso_cv = LassoCV(cv = 5, verbose=2)
```

```python
lasso_cv
```

```python
lasso_cv.fit(X_train, y_train)
```

```python
y_pred = lasso_cv.predict(X_test)
score = r2_score(y_test, y_pred)
print(f"The R square of Lasso CV {score}")
```

```python
from sklearn.linear_model import RidgeCV
ridgecv = RidgeCV(cv = 5)
ridgecv.fit(X_train, y_train)
```

```python
y_pred = ridgecv.predict(X_test)
r2_score(y_test, y_pred)
```

```python
ridgecv.get_params()
```

```python
from sklearn.linear_model import ElasticNetCV
```

```python
elastic_cv = ElasticNetCV(cv = 5)
elastic_cv.fit(X_train, y_train)
```

```python
y_pred = elastic_cv.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

#CV with hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import Lasso, Ridge
```

```python
# define the model
lasso = Lasso()
```

```python
lasso
```

```python
# grid search cv >> Dictionary with parameters names (str) as keys and lists of parameter settings to try as values, or a list of such dictionaries, in which case the grids spanned by each dictionary in the list are explored. This enables searching over any sequence of parameter settings.
param_grid = {'alpha' : [0.001, 0.01, 0.1, 1, 10, 100]}
```

```python
param_grid
```

```python
grid_search = GridSearchCV(estimator=lasso, param_grid=param_grid, cv = 5, scoring = 'r2', verbose =2)
```

```python
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_params_['alpha']
```

```python
grid_search.best_score_ # mean cross-validated score
```

```python
grid_search.best_estimator_ #highest score
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
# randomized search cv fo lasso
lasso = Lasso()
param_dist = {'alpha' : [0.001, 0.01, 0.1, 10, 1, 100]}
randomized_search_cv = RandomizedSearchCV(estimator=lasso, param_distributions=param_dist, cv = 5, scoring = 'r2', verbose = 2, n_iter=3)
```

```python
randomized_search_cv
```

```python
randomized_search_cv.fit(X_train, y_train)
```

```python
# due to n_itern random 3 parameters are selected
```

```python
randomized_search_cv.best_params_
```

```python
randomized_search_cv.best_estimator_
```

```python
randomized_search_cv.best_score_
```

```python
y_pred = randomized_search_cv.best_estimator_.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

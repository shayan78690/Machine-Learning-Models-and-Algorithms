```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=10, n_redundant=5, n_informative=5, n_classes=2, random_state=1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
# hyperparameter tuning
# Grid Search CV
```

```python
from sklearn.model_selection import GridSearchCV
params = {'penalty' : ('l1', 'l2', 'elasticnet'), 'C' : [1, 2, 10, 20, 30, 40]}
```

```python
params
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
classifier
```

```python
clf = GridSearchCV(classifier, param_grid=params, cv = 5, verbose = 2)
```

```python
clf
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
clf.best_estimator_
```

```python
# now prediction two ways directly predict from best_estimator or build a seperate model
clf.best_estimator_.predict(X_test)
```

```python
model = LogisticRegression(C = 1, penalty='l2')
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
model.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

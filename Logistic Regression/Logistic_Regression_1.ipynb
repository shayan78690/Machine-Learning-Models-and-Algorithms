# Splitting Data

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Scaling (Optional)

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# logistic regression gives you the probability
```

```python
classifier.predict_proba(X_test) # corresponding to higer probability class is predicted, the cutoff bydefault is 0.5
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# first two rows, 0 and 1 >> precision recall, fi score and no of data points, with respect to respective class
# macroavg >> for each class take simple avg >> avg precision of class 0 and class 1 will be macroavg
# similarly for recall, fi score
# support will be total no of datapoints
# weighted avg is used in case of class imbalance, it gives importance to performance of weighted class
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
# get predicted probability of class 1
y_pred_proba = classifier.predict_proba(X_test)[:, 1] # probability for 1 class
```

```python
y_pred_proba
```

```python
# receiver operation characteristics curve
# input : y_test, yred_proba
# it will return TPR, FPR with different cutoff of probability
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```

```python
thresholds
```

```python
fpr
```

```python
tpr
```

```python
auc(fpr, tpr) # because it is a 100% accurate model
```

```python
roc_auc = auc(fpr, tpr)
```

```python
plt.figure(figsize = (8, 6)) # Creat a new figure with a specified value
plt.plot(fpr, tpr, color = 'darkorange', linewidth=2, label='ROC curve (area = %0.2f) ' % roc_auc) # Plot ROC curve
plt.plot([0, 1], [0, 1], color = 'navy', linewidth = 2, linestyle = '--') # plot the diagonal line representing random guessing
plt.xlim([0.0, 1.0]) # set the x-axis limits
plt.ylim([0.0, 1.05]) # set y-axis label
plt.title("Receiver Operating Characteristics (ROC) curve") # set the plot
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc = "lower right")
plt.show()
```

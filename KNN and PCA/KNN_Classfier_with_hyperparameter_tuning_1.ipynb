```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=3, n_redundant=0, n_classes=2, random_state=1)
```

```python
X.shape
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_neighbors': [3, 5, 6, 7, 9, 11, 13],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size': [20, 30, 40, 50]
}
```

```python
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv = 5, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
#predict with the best esitimator

best_model = grid.best_estimator_
y_pred_tuned = best_model.predict(X_test)
y_pred_tuned
```

```python
# Evaluate the tuned model
print("Tuned model performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_tuned)}")
print("Classification Report:")
print(classification_report(y_test, y_pred_tuned))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_tuned))
```

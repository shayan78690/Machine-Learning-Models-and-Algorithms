```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# gradient boosting classifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X, y = make_classification(n_samples = 1000, n_features=20, n_classes=2, random_state=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
classifier = GradientBoostingClassifier()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param_grid = {
    'learning_rate': [0.01, 0.1, 0.05, 0.2],
    'n_estimators': [100, 200, 300]
}
```

```python
gbc = GradientBoostingClassifier()
grid = GridSearchCV(estimator=gbc, param_grid=param_grid, cv=5, verbose = 3)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
best_model = grid.best_estimator_
best_model
```

```python
y_pred = best_model.predict(X_test)
y_pred
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# gradient boosting regressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
X, y = make_regression(n_samples = 1000, n_features= 2, noise = 10, random_state = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
regressor = GradientBoostingRegressor()
regressor
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

```python
# hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param_grid = {
    'learning_rate': [0.01, 0.1, 0.05, 0.2],
    'n_estimators': [100, 200, 300]
}
```

```python
model = GradientBoostingRegressor()
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, verbose = 3)
```

```python
grid.fit(X_train, y_train)
```

```python
best_model = grid.best_estimator_
best_model
```

```python
y_pred = best_model.predict(X_test)
y_pred
```

```python
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

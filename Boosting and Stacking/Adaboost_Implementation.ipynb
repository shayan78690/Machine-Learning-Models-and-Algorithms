```python
# adaboost classifier
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state = 1)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
classifier = AdaBoostClassifier()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
print("Current model performance:")
print(f"Accuracy : {accuracy_score(y_test, y_pred)}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_gird = {
    'n_estimators' : [50, 100, 200],
    'learning_rate' : [0.01, 0.1, 1.0, 1.5, 2.0],
    'algorithm' : ['SAMME']
}
```

```python
ada = AdaBoostClassifier()
clf = GridSearchCV(estimator=ada, param_grid=param_gird, cv = 5, verbose = 3)
```

```python
clf
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_estimator_
```

```python
best_model = clf.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
print("Current model performance:")
print(f"Accuracy : {accuracy_score(y_test, y_pred)}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# adaboost regression
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
X, y = make_regression(n_samples=1000, n_features=2, noise = 10, random_state = 1)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
regressor = AdaBoostRegressor()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
print("Current model performance:")
print(f"R2 score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
```

```python
from sklearn.model_selection import GridSearchCV
param_gird = {
    'n_estimators' : [50, 100, 200],
    'learning_rate' : [0.01, 0.1, 0.5, 1.0, 1.5],
    'loss' : ['linear', 'square', 'exponential']
}
```

```python
grid_search = GridSearchCV(estimator=regressor, param_grid=param_gird, cv = 5, verbose = 3)
```

```python
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best = grid_search.best_estimator_
```

```python
y_pred = best.predict(X_test)
```

```python
print("Current model performance:")
print(f"R2 score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
```

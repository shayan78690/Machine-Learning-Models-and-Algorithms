```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons # creat non linear data

X, y = make_moons(n_samples=250, noise = 0.10)
```

```python
X
```

```python
y
```

```python
# y will not be used in modelling
```

```python
plt.scatter(X[:,0], X[:,1], c = y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
dbscan = DBSCAN(eps = 0.2) # change the value of epsilon
```

```python
dbscan
```

```python
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c = dbscan.labels_)
```

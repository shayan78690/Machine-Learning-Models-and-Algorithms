```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
X = data.data
y = data.target
```

```python
X
```

```python
y
```

```python
X.shape, y.shape
```

```python
# this is unsupervised learnign algorithm problem therfore y will not be used
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# we want to reduce the dimension in this scaled data
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca
```

```python
pca_scaled = pca.fit_transform(X_scaled)
```

```python
pca_scaled
```

```python
# visualize this
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c = y)
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1])
```

```python
# agglomerative clustering >> make dendogram
import scipy.cluster.hierarchy as sc
plt.figure(figsize = (20, 7))
plt.title("Dendogram")

sc.dendrogram(sc.linkage(pca_scaled, method= 'ward'))
plt.xlabel("Sample Data")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=2, metric = 'euclidean', linkage='ward')
```

```python
agg
```

```python
agg.fit(pca_scaled)
```

```python
agg.labels_
```

```python
# visulaize
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c = agg.labels_)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(pca_scaled, agg.labels_)
```

```python
silhouette_coeff = []
for k in range(2, 11):
  model = AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage = 'ward')
  model.fit(X_scaled)
  score = silhouette_score(X_scaled, model.labels_)
  silhouette_coeff.append(score)
```
